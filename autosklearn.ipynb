```
import sklearn
import autosklearn.classification as classifier
from sklearn import datasets
from sklearn.model_selection import train_test_split
```

```
X,y= sklearn.datasets.load_iris(return_X_y= True)

X_train,X_test,y_train,y_test= train_test_split(X,y,random_state= 0)
```

```
##Creating and aslo training the model on the iris dataset
autoclassifier= classifier.AutoSklearnClassifier(time_left_for_this_task= 180,per_run_time_limit= 40)
autoclassifier.fit(X_train,y_train)
```

```
from sklearn.metrics import accuracy_score
y_pred= autoclassifier.predict(X_test) 
```

```
score= accuracy_score(y_test,y_pred)
print(score)
```
